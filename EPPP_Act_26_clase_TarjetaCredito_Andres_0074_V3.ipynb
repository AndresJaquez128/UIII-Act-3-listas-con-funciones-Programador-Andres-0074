```python
# Programa realizado por Andres Jaquez NL = 27
# Fecha 11/05/2026
# Operaciones sobre listas utilizando funciones y menú con while True

# Lista donde se almacenarán los programadores
programadores = []


# Función para agregar un programador
def agregarprogramador():

    # Se solicita el nombre del programador
    nombre = input("Ingresa el nombre del programador: ")

    # El nombre se agrega a la lista
    programadores.append(nombre)

    # Mensaje de confirmación
    print("Programador agregado correctamente")


# Función para leer un programador por posición
def leerprogramador():

    # Se solicita la posición
    posicion = int(input("Ingresa la posición del programador: "))

    # Se verifica si la posición existe
    if posicion >= 0 and posicion < len(programadores):

        # Se muestra el programador encontrado
        print("Programador encontrado:", programadores[posicion])

    else:

        # Mensaje de error
        print("Posición no válida")


# Función para mostrar todos los programadores
def listarprogramador():

    # Se verifica si la lista tiene elementos
    if len(programadores) == 0:

        print("No hay programadores registrados")

    else:

        # Variable contador
        contador = 0

        # Ciclo while para recorrer la lista
        while contador < len(programadores):

            # Se muestra cada programador
            print(contador, "-", programadores[contador])

            # El contador aumenta
            contador = contador + 1


# Función para buscar un programador
def buscarprogramador():

    # Se solicita el nombre a buscar
    nombre = input("Ingresa el nombre a buscar: ")

    # Variable contador
    contador = 0

    # Variable para verificar si existe
    encontrado = False

    # Ciclo while para recorrer la lista
    while contador < len(programadores):

        # Se compara el nombre
        if programadores[contador] == nombre:

            # Se muestra la posición encontrada
            print("Programador encontrado en la posición:", contador)

            # La variable cambia a verdadero
            encontrado = True

        # El contador aumenta
        contador = contador + 1

    # Si no se encontró el nombre
    if encontrado == False:

        print("Programador no encontrado")


# Función para modificar un programador
def modificarprogramador():

    # Se solicita la posición
    posicion = int(input("Ingresa la posición a modificar: "))

    # Se verifica si la posición existe
    if posicion >= 0 and posicion < len(programadores):

        # Se solicita el nuevo nombre
        nuevo = input("Ingresa el nuevo nombre: ")

        # Se reemplaza el valor anterior
        programadores[posicion] = nuevo

        # Mensaje de confirmación
        print("Programador modificado correctamente")

    else:

        # Mensaje de error
        print("Posición no válida")


# Función para borrar un programador
def borrarprogramador():

    # Se solicita la posición
    posicion = int(input("Ingresa la posición a borrar: "))

    # Se verifica si la posición existe
    if posicion >= 0 and posicion < len(programadores):

        # Se elimina el programador
        programadores.pop(posicion)

        # Mensaje de confirmación
        print("Programador eliminado correctamente")

    else:

        # Mensaje de error
        print("Posición no válida")


# Menú principal
while True:

    print("\n--- MENÚ DE PROGRAMADORES ---")
    print("1. Agregar programador")
    print("2. Leer programador")
    print("3. Listar programadores")
    print("4. Buscar programador")
    print("5. Modificar programador")
    print("6. Borrar programador")
    print("7. Salir")

    # Se solicita una opción
    opcion = int(input("Selecciona una opción: "))

    # Se ejecuta la opción elegida
    if opcion == 1:

        agregarprogramador()

    elif opcion == 2:

        leerprogramador()

    elif opcion == 3:

        listarprogramador()

    elif opcion == 4:

        buscarprogramador()

    elif opcion == 5:

        modificarprogramador()

    elif opcion == 6:

        borrarprogramador()

    elif opcion == 7:

        print("Fin del programa")
        break

    else:

        print("Opción no válida")

print("Software desarrollado por Jaquez Andres NL = 27")
```
